# Linear Regression

## Simple Linear Regression

A straightforward approach for predicting a quantitaitve response $Y$ on the basis of a single predictor variable $X$.

$$Y \approx \underbrace{\beta_0}_{\text{intercept}} + \underbrace{\beta_1}_{\text{slope}} X$$

- **Intercept**: the expected value of $Y$ when $X = 0$
- **Slope**: the average increase in $Y$ associated with 1-unit increase in $X$

When the training data produces the estimates for the model coefficients, predicting X can be computed by:

$$\hat{y} = \hat{\beta_0} + \hat{\beta_1}x$$

### Estimating the Coefficients

---
> **Goal**: Find an intercept $\hat{\beta_0}$ and $\hat{\beta_1}$ such that the resulting line is as close as possible to $n$ data points.
---

*How do we measure closeness?*

- **Least Squares**: a statistical procedure to find the best fit for a set of data points by minimizing the sum of the offsets or residuals of points from the plotted curve

The *fit* is found by minimizing the sum of squared errors.

In the image above, the grey line represents an error.

**Residual Sum of Squares (RSS)**: measures the level of variance in the error term/residuals of a regression model
- The smaller the residual sum of squares, the better the model fits the data
- **Residual**: the difference between the predicted and the observed value of $y$
    - $e_i = y_i - \hat{y_i}$


$$ RSS = (y_1 - \hat{\beta_0} - \hat{\beta_1}x_1) ^ 2 + ... + (y_n - \hat{\beta_0} - \hat{\beta_1}x_n) $$

### Asessing the Accuracy of the Coefficient Estimates


We assume that there's a true relationship between $X$ and $Y$ ($Y = f(X) + \epsilon$). 

$f$ is approximated by a linear function $Y = \beta_0 + \beta_1 X + \epsilon$
- $\epsilon$: other variables that causes the variation in $Y$ (independent from $X$)

**Population Regression Line**: the best linear approximation to the true relationship between $X$ and $Y^1$

*The main difference between the population & least squares regression lines comes down to using info from a sample to estimate charasteristics of the population (think of $\hat{\mu}$ &rarr; $\mu$)*

In other words, the unknown coefficients, $\beta_0$ and $\beta_1$, are estimated using $\hat{\beta_0}$ and $\hat{\beta_1}$.
